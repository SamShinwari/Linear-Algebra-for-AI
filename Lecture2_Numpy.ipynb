## NumPy Basics

NumPy (Numerical Python) is a powerful library for numerical operations in Python. It provides support for large, multi-dimensional arrays and matrices, along with a collection of high-level mathematical functions to operate on these arrays. It's a fundamental package for scientific computing with Python.

### Creating Arrays

NumPy arrays can be created from Python lists or tuples. You can create 1D, 2D, or N-dimensional arrays.

```python
import numpy as np

a = np.array([1,2,3])
b = np.array([[1,2],[3,4]])

print(f"1D Array: {a}")
print(f"2D Array:\n {b}")
```

### Checking Shape

The `shape` attribute returns a tuple of integers indicating the size of the array in each dimension. For a 1D array, it returns a tuple with one element. For a 2D array, it returns `(rows, columns)`.

```python
print(f"Shape of 1D array (a): {a.shape}")
print(f"Shape of 2D array (b): {b.shape}")
```

### Array Dimensions

The `ndim` attribute returns the number of dimensions of the array. A 1D array has `ndim = 1`, a 2D array has `ndim = 2`, and so on.

```python
print(f"Dimensions of 1D array (a): {a.ndim}")
print(f"Dimensions of 2D array (b): {b.ndim}")
```

### Size

The `size` attribute returns the total number of elements in the array.

```python
print(f"Total elements in array (b): {b.size}")
```

### Data Type

The `dtype` attribute returns the data type of the elements in the array. NumPy automatically infers the data type during array creation, but you can also specify it explicitly.

```python
print(f"Data type of elements in array (b): {b.dtype}")
```

### Creating Special Arrays

NumPy provides functions to create arrays with pre-filled content, such as arrays of zeros, ones, or identity matrices.

```python
print(f"Zeros array (3x3):\n {np.zeros((3,3))}")
print(f"Ones array (2x4):\n {np.ones((2,4))}")
print(f"Identity matrix (4x4):\n {np.eye(4)}")
print(f"Random array (3x3):\n {np.random.rand(3,3)}")
```

### Indexing

You can access individual elements of a NumPy array using indexing, similar to Python lists. For multi-dimensional arrays, you use a comma-separated tuple of indices `[row, column]`.

```python
A_idx = np.array([[5,6,7],
              [8,9,10]])

print(f"Original array A_idx:\n {A_idx}")
print(f"Element at (0,1): {A_idx[0,1]}")
```

### Slicing

Slicing allows you to extract sub-arrays. You can specify a range of indices using the colon operator `:`. For multi-dimensional arrays, you can slice along each dimension.

```python
print(f"Original array A_idx:\n {A_idx}")
print(f"Second column of A_idx: {A_idx[:,1]}")
```

### Reshaping

The `reshape()` method allows you to change the shape of an array without changing its data. The new shape must have the same total number of elements as the original array.

```python
A_res = np.arange(12)
print(f"Original 1D array A_res: {A_res}")
print(f"Reshaped A_res to (3,4):\n {A_res.reshape(3,4)}")
```

### Flatten

The `flatten()` method returns a copy of the array collapsed into one dimension.

```python
B_flat = np.array([[1,2],[3,4]])
print(f"Original 2D array B_flat:\n {B_flat}")
print(f"Flattened B_flat: {B_flat.flatten()}")
```
