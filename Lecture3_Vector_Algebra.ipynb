# Lecture 1: Introduction to Vectors

## What is a Vector?

A **vector** is an ordered collection of numbers.

Unlike a scalar (single value), a vector contains multiple values.

Example

$$
v=\begin{bmatrix}
2\\
5\\
7
\end{bmatrix}
$$

This vector has **three elements**, so it is called a **3-dimensional vector (3D vector).**

---

## Real-world Example

Suppose you want to describe a student.

| Feature | Value  |
| ------- | ------ |
| Age     | 20     |
| Height  | 175 cm |
| Weight  | 68 kg  |

Instead of writing these separately, we combine them into one vector:

$$
Student=
\begin{bmatrix}
20\\
175\\
68
\end{bmatrix}
$$

Machine Learning treats every data sample as a vector.

$$

# Lecture 2: Vector Norms

## What is Magnitude?

Magnitude tells **how long a vector is**.

Example

Suppose

$$
v=
\begin{bmatrix}
3\\
4
\end{bmatrix}
$$

The length is

$$
\sqrt{3^2+4^2}=5
$$

This comes from the Pythagorean theorem.

---

## L1 Norm

Also called **Manhattan Distance**.

Formula

$$
||v||_1=\sum |x_i|
$$

Example

$$
\begin{bmatrix}
3\\
-5\\
2
\end{bmatrix}
$$

L1 Norm

$$
3+5+2=10
$$

Applications

* Sparse models
* LASSO Regression
* Feature selection

---

## L2 Norm

Also called Euclidean Norm.

Formula

$$
||v||_2=
\sqrt{\sum x_i^2}
$$

Example

$$
\begin{bmatrix}
3\\
4
\end{bmatrix}
$$

Result

5

Applications

* Deep Learning
* Optimization
* Weight regularization

---

## Infinity Norm

Largest absolute value.

Example

$$
\begin{bmatrix}
4\\
8\\
-2
\end{bmatrix}
$$

Infinity norm

8

Applications

* Maximum error analysis

# Dot Product

The **dot product** multiplies corresponding elements of two vectors and then sums the results.

Formula

$$
A \cdot B = \sum_{i=1}^{n} A_i B_i
$$

Example

$$
A=
\begin{bmatrix}
2\\
3
\end{bmatrix},
\quad
B=
\begin{bmatrix}
4\\
5
\end{bmatrix}
$$

Calculation

$$
(2 \times 4) + (3 \times 5) = 8 + 15 = 23
$$

Python

````python
A = np.array([2,3])
B = np.array([4,5])

print(np.dot(A,B))
````

Output

```
23
```

### Why is the Dot Product Important?

* Measures how much two vectors point in the same direction.
* If the dot product is **positive**, the vectors point roughly in the same direction.
* If it is **zero**, the vectors are perpendicular (orthogonal).
* If it is **negative**, they point in opposite directions.

In AI:

*   **Neural networks compute weighted sums using dot products.** This is fundamental for how neurons process inputs.
*   **Attention mechanisms in Transformers use dot products to compare queries and keys.** This determines how much focus different parts of a sequence receive.
*   **Recommendation systems compare user and item vectors.** For instance, if a user vector represents their preferences and an item vector represents item features, a high dot product suggests a good match.
*   **Feature Matching:** In computer vision, a dot product can compare two feature vectors extracted from images to see how similar they are, for example, to identify objects.

# Cosine Similarity

Cosine similarity measures the **angle** between two vectors rather than their lengths.

Formula

$$
\text{Cosine Similarity} =
\frac{A \cdot B}{|A||B|}
$$

The result ranges from **-1 to 1**:

*   **1** → Exactly the same direction (very similar)
*   **0** → Unrelated (orthogonal)
*   **-1** → Opposite directions

### Example

Suppose:

*   Sentence 1: "I love machine learning."
*   Sentence 2: "Machine learning is amazing."

After converting them into embeddings (vectors), cosine similarity might be **0.95**, indicating that the meanings are very similar.

If Sentence 2 were "The weather is sunny today," the cosine similarity might be **0.10**, indicating little semantic similarity.

### More Practical Examples:

*   **Document Search:** When you type a query into a search engine, the query is converted into a vector, and then cosine similarity is used to find documents (also represented as vectors) that are most semantically similar to your query.
*   **Customer Support Chatbots:** If a user asks a question, the chatbot converts the question into a vector and compares it using cosine similarity to a database of known questions and their answers to provide the most relevant response.
*   **Plagiarism Detection:** Documents can be compared using cosine similarity to identify sections that are semantically very close, potentially indicating plagiarism.

### Python

````python
from numpy.linalg import norm
import numpy as np

A = np.array([2, 3])
B = np.array([4, 5])

cosine = np.dot(A, B) / (norm(A) * norm(B))
print(cosine)
````

## Python Example: Comparing Multiple Sentences with Cosine Similarity

To practically apply cosine similarity to text, we first need to convert our sentences into numerical vector representations. A common method for this is TF-IDF (Term Frequency-Inverse Document Frequency) vectorization. Then, we can calculate the similarity between these vectors.

````python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Define a set of sentences
sentences = [
    "The cat sat on the mat.",
    "The dog lay on the rug.",
    "A feline rested on the carpet.",
    "Machine learning is a fascinating field.",
    "The weather is sunny today."
]

# 1. Initialize TF-IDF Vectorizer
# This converts text documents into a matrix of TF-IDF features.
vectorizer = TfidfVectorizer()

# 2. Fit and transform the sentences to get TF-IDF vectors
tfidf_matrix = vectorizer.fit_transform(sentences)

# 3. Calculate the cosine similarity matrix
# This computes the pairwise cosine similarity between all sentence vectors.
cosine_sim_matrix = cosine_similarity(tfidf_matrix)

# 4. Display the results in a readable format (e.g., a DataFrame)
sim_df = pd.DataFrame(cosine_sim_matrix, index=sentences, columns=sentences)
print("Cosine Similarity Matrix:")
display(sim_df)

print("\nInterpretation:")
print("- Values close to 1 indicate high similarity (e.g., 'The cat sat on the mat.' and 'A feline rested on the carpet.').")
print("- Values close to 0 indicate low similarity (e.g., 'The cat sat on the mat.' and 'The weather is sunny today.').")
````
