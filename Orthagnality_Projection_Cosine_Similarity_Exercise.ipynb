# Programming Lab

Write Python programs to:

1. Compute the angle between two vectors.
2. Check whether two vectors are orthogonal.
3. Compute the projection of one vector onto another.
4. Normalize vectors to unit length.
5. Compute cosine similarity between two sentence embeddings.

```python
# 1. Compute the angle between two vectors
import numpy as np

def angle_between_vectors(vec1, vec2):
    dot_product = np.dot(vec1, vec2)
    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)
    if norm_vec1 == 0 or norm_vec2 == 0:
        return 0 # Or handle as an error/special case
    cosine_theta = dot_product / (norm_vec1 * norm_vec2)
    # Ensure cosine_theta is within valid range [-1, 1] to avoid arccos errors
    cosine_theta = np.clip(cosine_theta, -1.0, 1.0)
    angle_rad = np.arccos(cosine_theta)
    return np.degrees(angle_rad)

# Example usage:
A = np.array([3, 4])
B = np.array([4, 3])
print(f"Angle between A and B: {angle_between_vectors(A, B):.2f} degrees\n")


# 2. Check whether two vectors are orthogonal
def are_orthogonal(vec1, vec2, tolerance=1e-9):
    dot_product = np.dot(vec1, vec2)
    return abs(dot_product) < tolerance

# Example usage:
C = np.array([1, 0])
D = np.array([0, 1])
E = np.array([1, 1])
print(f"Are C and D orthogonal? {are_orthogonal(C, D)}")
print(f"Are C and E orthogonal? {are_orthogonal(C, E)}\n")


# 3. Compute the projection of one vector onto another
def vector_projection(A_vec, B_vec):
    if np.linalg.norm(B_vec) == 0:
        return np.zeros_like(A_vec) # Projection onto a zero vector is zero vector
    dot_product = np.dot(A_vec, B_vec)
    magnitude_B_squared = np.dot(B_vec, B_vec)
    projection = (dot_product / magnitude_B_squared) * B_vec
    return projection

# Example usage:
F = np.array([4, 2])
G = np.array([1, 0])
print(f"Projection of F onto G: {vector_projection(F, G)}\n")


# 4. Normalize vectors to unit length
def normalize_vector(vec):
    norm = np.linalg.norm(vec)
    if norm == 0:
        return vec # Handle zero vector
    return vec / norm

# Example usage:
H = np.array([3, 4])
print(f"Normalized H: {normalize_vector(H)}")
print(f"Norm of Normalized H: {np.linalg.norm(normalize_vector(H)):.2f}\n")


# 5. Compute cosine similarity between two sentence embeddings
def cosine_similarity(embedding1, embedding2):
    dot_product = np.dot(embedding1, embedding2)
    norm_emb1 = np.linalg.norm(embedding1)
    norm_emb2 = np.linalg.norm(embedding2)
    if norm_emb1 == 0 or norm_emb2 == 0:
        return 0 # Handle zero embeddings
    return dot_product / (norm_emb1 * norm_emb2)

# Example usage (illustrative embeddings):
sentence_embedding_1 = np.array([0.2, 0.8, 0.1, 0.5])
sentence_embedding_2 = np.array([0.3, 0.7, 0.2, 0.4])
sentence_embedding_3 = np.array([-0.5, 0.1, 0.9, -0.2])

print(f"Cosine similarity between emb1 and emb2: {cosine_similarity(sentence_embedding_1, sentence_embedding_2):.4f}")
print(f"Cosine similarity between emb1 and emb3: {cosine_similarity(sentence_embedding_1, sentence_embedding_3):.4f}")
```
