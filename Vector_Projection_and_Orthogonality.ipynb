### 2. Angle Between Two Vectors
The angle between two vectors tells us how similar their directions are.
Imagine two arrows starting from the same point:
* Small angle → Similar direction
* 90° angle → Completely independent
* 180° angle → Opposite direction

The angle is found using the dot product:

$$A \cdot B = |A| |B| \cos(\theta)$$

Rearranging,

$$\cos(\theta)=\frac{A \cdot B}{|A| |B|}$$

This is the same formula used in cosine similarity.

#### Example
Let

$$A=\begin{bmatrix}3\\4\end{bmatrix}$$

$$B=\begin{bmatrix}4\\3\end{bmatrix}$$

**Step 1: Dot Product**

$$A\cdot B=(3\times4)+(4\times3)=24$$

**Step 2: Magnitudes**

$$||A||=5$$

$$||B||=5$$

**Step 3**

$$\cos(\theta)=\frac{24}{25}=0.96$$

**Step 4**

$$\theta=\cos^{-1}(0.96)$$

$$\theta\approx16.26^\circ$$

The vectors point in nearly the same direction.

### 3. Orthogonal Vectors
Two vectors are orthogonal if they are perpendicular.
Angle between them:

$$90^\circ$$

A very important property:
If vectors are orthogonal,

$$A\cdot B=0$$

#### Example

$$A=\begin{bmatrix}1\\0\end{bmatrix}$$

$$B=\begin{bmatrix}0\\1\end{bmatrix}$$

**Dot Product**

$$1\times0+0\times1=0$$

Therefore, A and B are orthogonal.

#### Why is Orthogonality Important?
Orthogonal vectors carry independent information.
Imagine two exam subjects:
* Mathematics
* Literature
Good performance in one does not necessarily predict performance in the other. Their feature vectors may be close to orthogonal, indicating they capture different aspects of a student's abilities.

### 4. Orthonormal Vectors
An orthonormal vector satisfies two conditions:
1. Magnitude = 1
2. Orthogonal to the other vectors

#### Example

$$u_1=\begin{bmatrix}1\\0\end{bmatrix}$$

$$u_2=\begin{bmatrix}0\\1\end{bmatrix}$$

These vectors are orthonormal because:
* Their lengths are 1.
* Their dot product is 0.

#### Applications
* Principal Component Analysis (PCA)
* QR Decomposition
* Singular Value Decomposition (SVD)
* Transformer models
* Signal Processing

### 5. Vector Projection
Projection answers the question:
How much of one vector lies in the direction of another vector?
Imagine shining a flashlight on vector A so that its shadow falls along vector B. The shadow is the projection of A onto B.

#### Formula
The projection of vector (A) onto vector (B) is:

$$\text{Proj}_B(A)=\frac{A\cdot B}{||B||^2}B$$

#### Example
Suppose

$$A=\begin{bmatrix}4\\2\end{bmatrix}$$

$$B=\begin{bmatrix}1\\0\end{bmatrix}$$

**Step 1**
Dot product

$$A\cdot B=4$$

**Step 2**

$$||B||^2=1$$

**Step 3**

$$\text{Projection} \; 4 \begin{bmatrix}1\\0\end{bmatrix} \begin{bmatrix}4\\0\end{bmatrix}$$

The projection keeps only the component of A that points in the direction of B.

### 10. Applications in Transformers and LLMs
In Transformer models, each token is represented by three vectors:
* Query (Q)
* Key (K)
* Value (V)

The attention mechanism compares queries and keys using the dot product:

$$\text{Score}=Q\cdot K$$

A larger dot product means the vectors point in more similar directions, so the model pays more attention to that token.

The complete scaled dot-product attention formula is:

$$\text{Attention}(Q,K,V)
\text{Softmax}
\left(
\frac{QK^T}{\sqrt{d_k}}
\right)
V$$

Here:
* $(QK^T)$ computes similarity scores.
* $(\sqrt{d_k})$ prevents the scores from becoming too large as the embedding dimension grows.
* Softmax converts the scores into probabilities.
* These probabilities determine how much each value vector contributes to the final output.

This is one of the key mathematical ideas behind modern LLMs such as GPT, Llama, and Gemma.



#### Python Example: Calculating the angle between two vectors

```python
import numpy as np

A = np.array([3,4])
B = np.array([4,3])

theta = np.arccos(
    np.dot(A,B)/(np.linalg.norm(A)*np.linalg.norm(B))
)

print(np.degrees(theta))
# Output: 16.26
```

#### Python Example: Calculating vector projection

```python
import numpy as np

A = np.array([4,2])
B = np.array([1,0])

projection = (np.dot(A,B)/np.dot(B,B))*B

print(projection)
# Output: [4. 0.]
```

1. What is the difference between a dot product and cosine similarity?

**Answer to 1:**

*   **Dot Product:** The dot product (or scalar product) is an algebraic operation that takes two equal-length sequences of numbers (usually coordinate vectors) and returns a single number. Geometrically, it is the product of the Euclidean magnitudes of the two vectors and the cosine of the angle between them. It indicates the degree to which two vectors point in the same direction.

*   **Cosine Similarity:** Cosine similarity is a measure of similarity between two non-zero vectors of an inner product space that measures the cosine of the angle between them. It is essentially the dot product normalized by the magnitudes of the two vectors. It ranges from -1 (completely opposite) to 1 (exactly the same direction), with 0 indicating orthogonality (independence). Cosine similarity is a normalized version of the dot product, making it suitable for comparing the *direction* rather than the magnitude of vectors.

2. Why do orthogonal vectors represent independent information?

**Answer to 2:**

Orthogonal vectors represent independent information because their dot product is zero ($A \cdot B = 0$), which implies that the angle between them is 90 degrees ($cos(90^\circ) = 0$). This means that neither vector contributes any component to the direction of the other. In a data context, if two feature vectors are orthogonal, knowing the value of one feature gives you no information about the value of the other feature. They capture distinct and non-overlapping aspects of the data.
