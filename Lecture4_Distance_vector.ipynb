# 1. What is Distance?

In mathematics, **distance** measures how far two points or vectors are from each other.

Think about Google Maps.

Suppose your home is at one location and your university is at another.

The distance tells you **how far apart they are**.

Similarly, in Machine Learning, every data sample is represented as a vector.

Example:

Student A

[
A=
\begin{bmatrix}
20\n170\n65
\end{bmatrix}
]

Student B

[
B=
\begin{bmatrix}
22\n175\n68
\end{bmatrix}
]

The question is:

**How similar are these students?**

To answer this, we calculate the **distance** between their vectors.

# 2. Euclidean Distance

This is the **straight-line distance** between two points.

It comes directly from the **Pythagorean Theorem**.

Formula

[
d(A,B)=
\sqrt{
(x_1-y_1)^2+
(x_2-y_2)^2+
...
+(x_n-y_n)^2
}
]

---

## Euclidean Distance Example

Suppose we have two vectors A and B. We will calculate the Euclidean distance between them.

```python
import numpy as np

A = np.array([2,3])
B = np.array([5,7])

distance = np.linalg.norm(A-B)

print(f"Euclidean Distance = {distance}")
```

## Euclidean Distance Applications

* Image Classification
* Face Recognition
* KNN
* K-Means
* Robotics

# 3. Manhattan Distance

Imagine you are driving in a city with square blocks. You **cannot drive diagonally**.

You must move:

* Left
* Right
* Up
* Down

This is Manhattan Distance.

Formula

[
d=
|x_1-y_1|
+
|x_2-y_2|
+\cdots
+
|x_n-y_n|
]

---

## Manhattan Distance Example

Using the same vectors A and B, we will calculate the Manhattan distance.

```python
import numpy as np

A = np.array([2,3])
B = np.array([5,7])

distance = np.sum(np.abs(A-B))

print(f"Manhattan Distance = {distance}")
```

## Manhattan Distance Applications

* Grid navigation
* Sparse datasets
* Feature selection
* L1 Regularization

---

# 4. Chebyshev Distance

Chebyshev Distance measures the **largest difference** along any single dimension.

Formula

[
d=
\max
(|x_i-y_i|)
]

---

## Chebyshev Distance Example

Again, using vectors A and B, let's calculate the Chebyshev distance.

```python
import numpy as np

A = np.array([2,3])
B = np.array([5,7])

distance = np.max(np.abs(A-B))

print(f"Chebyshev Distance = {distance}")
```

## Chebyshev Distance Applications

* Chess programming
* Robotics
* Industrial quality control

# Mini Project

Build a simple **Student Similarity Finder**:

1. Create a dataset of 10 students with marks in 5 subjects.
2. Represent each student as a vector.
3. Compute Euclidean and Manhattan distances between students.
4. Identify the most similar student to a selected student.

## Mini Project: Step 1 - Create a dataset of 10 students with marks in 5 subjects.

We will create a synthetic dataset using NumPy. Each row will represent a student, and each column will represent marks in a subject (e.g., Math, Physics, Chemistry, Biology, English). Marks will be random integers between 40 and 100.

```python
import numpy as np
import pandas as pd

# Set a random seed for reproducibility
np.random.seed(42)

# Create a dataset of 10 students, each with marks in 5 subjects
num_students = 10
num_subjects = 5

# Generate random marks between 40 and 100
student_marks = np.random.randint(40, 101, size=(num_students, num_subjects))

# Create a DataFrame for better visualization
student_names = [f'Student_{i+1}' for i in range(num_students)]
subject_names = [f'Subject_{i+1}' for i in range(num_subjects)]
df_students = pd.DataFrame(student_marks, index=student_names, columns=subject_names)

print("Student Dataset (Marks in 5 Subjects):")
display(df_students)
```

## Mini Project: Step 2 - Represent each student as a vector.

Our DataFrame `df_students` already represents each student as a row vector, where the marks in 5 subjects form the dimensions of the vector. We can directly use these rows (or their underlying NumPy arrays) for distance calculations.

## Mini Project: Step 3 - Compute Euclidean and Manhattan distances between students.

We will calculate the pairwise distances between all students using both Euclidean and Manhattan metrics. This will help us understand how similar each student is to every other student.

```python
from scipy.spatial.distance import pdist, squareform

# Extract the marks as a NumPy array for distance calculation
student_vectors = df_students.values

# Calculate pairwise Euclidean distances
euclidean_distances = pdist(student_vectors, metric='euclidean')
euclidean_distance_matrix = pd.DataFrame(squareform(euclidean_distances), index=student_names, columns=student_names)

print("Pairwise Euclidean Distances between Students:")
display(euclidean_distance_matrix)

# Calculate pairwise Manhattan distances
manhattan_distances = pdist(student_vectors, metric='cityblock') # 'cityblock' is the metric name for Manhattan distance
manhattan_distance_matrix = pd.DataFrame(squareform(manhattan_distances), index=student_names, columns=student_names)

print("\nPairwise Manhattan Distances between Students:")
display(manhattan_distance_matrix)
```

## Mini Project: Step 4 - Identify the most similar student to a selected student.

Now, let's pick a specific student (e.g., 'Student_1') and find which other student is most similar to them based on their Euclidean distance. The most similar student will have the smallest non-zero distance.

```python
# Select a student to compare against
selected_student_name = 'Student_1'

# Get the row of distances for the selected student from the Euclidean distance matrix
distances_to_selected_student = euclidean_distance_matrix.loc[selected_student_name]

# Remove the distance to itself (which is 0)
distances_to_selected_student = distances_to_selected_student[distances_to_selected_student > 0]

# Find the student with the minimum distance
most_similar_student_name = distances_to_selected_student.idxmin()
min_distance = distances_to_selected_student.min()

print(f"The selected student is: {selected_student_name}")
print(f"The most similar student to {selected_student_name} (based on Euclidean distance) is: {most_similar_student_name}")
print(f"With a Euclidean distance of: {min_distance:.2f}")

# You can repeat this for Manhattan distance as well if desired:
# distances_to_selected_student_manhattan = manhattan_distance_matrix.loc[selected_student_name]
# distances_to_selected_student_manhattan = distances_to_selected_student_manhattan[distances_to_selected_student_manhattan > 0]
# most_similar_student_name_manhattan = distances_to_selected_student_manhattan.idxmin()
# min_distance_manhattan = distances_to_selected_student_manhattan.min()
# print(f"\nThe most similar student to {selected_student_name} (based on Manhattan distance) is: {most_similar_student_name_manhattan}")
# print(f"With a Manhattan distance of: {min_distance_manhattan}")
```

## Practice Question: Manhattan Distance Calculation

Now, let's calculate the Manhattan distance between the same points P1=(1, 2) and P2=(4, 6).

```python
import numpy as np

P1 = np.array([1, 2])
P2 = np.array([4, 6])

manhattan_distance = np.sum(np.abs(P1 - P2))

print(f"The Manhattan distance between (1,2) and (4,6) is: {manhattan_distance}")
```

## Practice Question: Chebyshev Distance Calculation

Finally, let's calculate the Chebyshev distance between P1=(1, 2) and P2=(4, 6).

```python
import numpy as np

P1 = np.array([1, 2])
P2 = np.array([4, 6])

chebyshev_distance = np.max(np.abs(P1 - P2))

print(f"The Chebyshev distance between (1,2) and (4,6) is: {chebyshev_distance}")
```

## Answers to Practice Questions

### 1. What is the difference between Euclidean and Manhattan distance?

*   **Euclidean Distance:** This is the straight-line distance between two points, calculated as the square root of the sum of the squared differences of their coordinates. It's like walking directly across a field.
*   **Manhattan Distance:** Also known as L1 distance or city block distance, this is the sum of the absolute differences of their coordinates. It's like navigating a city grid where you can only move along perpendicular streets (horizontal and vertical).

### 5. Which distance metric would you choose for:

*   **KNN on continuous features?**
    *   **Euclidean Distance:** Generally preferred for continuous features, especially when the magnitude of the difference between dimensions is important and features are on a similar scale. It assumes a straight-line relationship between points.

*   **Navigation on city streets?**
    *   **Manhattan Distance:** This is the most appropriate choice for city street navigation as it accurately reflects travel along a grid-like network where diagonal movement is not possible.

*   **Measuring the maximum deviation in sensor readings?**
    *   **Chebyshev Distance:** This metric is ideal for scenarios where the largest difference along any single dimension is of most concern, such as finding the maximum error in a multi-sensor system. It focuses on the 'worst-case' difference among dimensions.

# Key Takeaways

* A **distance metric** measures how different two vectors are.
* **Euclidean distance** gives the straight-line distance.
* **Manhattan distance** sums the absolute differences along each dimension.
* **Chebyshev distance** considers only the largest single-coordinate difference.
* Distance metrics are fundamental to **KNN**, **K-Means**, **recommendation systems**, **semantic search**, **embeddings**, and **LLMs**.
